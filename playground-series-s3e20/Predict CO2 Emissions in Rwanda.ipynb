### GroupBy Function
Let's do something with groupby function.
- Groupby by:
  
      - year
      - latitude
      - longitude
      - week no
  Here, emission is plotted with respect to these colums to know what is the distribution of C02 emission across these columns.

```python
df.groupby(df['year']).max()['emission'].plot(kind='bar');
```

```python
df.groupby(df['year']).mean()['emission'].plot(kind='bar');
```

```python
df.groupby(df['latitude']).mean()['emission'].plot(kind='line');
```

```python
df.groupby(df['longitude']).mean()['emission'].plot(kind='line');
```

```python
df.groupby(df['week_no']).mean()['emission'].plot(kind='line');
```
