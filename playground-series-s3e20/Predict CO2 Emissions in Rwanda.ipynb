### GroupBy Function
Let's do something with groupby function.
- Groupby by:
  
      - year
      - latitude
      - longitude
      - week no
  Here, emission is plotted with respect to these colums to know what is the distribution of C02 emission across these columns.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the relevant data

# Create a 3x2 grid of subplots
fig, axes = plt.subplots(3, 2, figsize=(12, 10))

# Plot maximum emission by year in the first subplot
df.groupby(df['year']).max()['emission'].plot(kind='bar', ax=axes[0, 0])
axes[0, 0].set_title('Max Emission by Year')

# Plot mean emission by year in the second subplot
df.groupby(df['year']).mean()['emission'].plot(kind='bar', ax=axes[0, 1])
axes[0, 1].set_title('Mean Emission by Year')

# Plot mean emission by latitude in the third subplot
df.groupby(df['latitude']).mean()['emission'].plot(kind='line', ax=axes[1, 0])
axes[1, 0].set_title('Mean Emission by Latitude')

# Plot mean emission by longitude in the fourth subplot
df.groupby(df['longitude']).mean()['emission'].plot(kind='line', ax=axes[1, 1])
axes[1, 1].set_title('Mean Emission by Longitude')

# Plot mean emission by week number in the fifth subplot
df.groupby(df['week_no']).mean()['emission'].plot(kind='line', ax=axes[2, 0])
axes[2, 0].set_title('Mean Emission by Week Number')

df.groupby(df['week_no']).max()['emission'].plot(kind='line', ax=axes[2, 1])
axes[2, 1].set_title('Max Emission by Week Number')
# Remove any empty subplots
fig.tight_layout()

# Show the plots
plt.show()
```
